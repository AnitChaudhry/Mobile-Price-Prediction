```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("/kaggle/input/mobile-price-classification/train.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
import missingno as msno
import matplotlib.pyplot as plt
msno.bar(data)
plt.show()
```

```python
corr=data.corr()
fig = plt.figure(figsize=(15,12))
r = sns.heatmap(corr, cmap='Purples',linecolor="black",linewidth=0)
r.set_title("Correlation ")
plt.show()
```

```python
sns.jointplot(x='ram',y='price_range',data=data,color='red',kind='kde');
```

```python
sns.jointplot(x='mobile_wt',y='price_range',data=data,kind='kde');
```

```python
sns.boxplot(x="price_range", y="battery_power", data=data)
plt.show()
```

```python
plt.figure(figsize=(10,6))
data['fc'].hist(alpha=0.5,color='blue',label='Front camera')
data['pc'].hist(alpha=0.5,color='red',label='Primary camera')
plt.legend()
plt.xlabel('MegaPixels')
```

```python
labels = data["price_range"].value_counts().index
sizes = data["price_range"].value_counts()
colors = ['green','pink','magenta','cyan',"orange","yellow"]
plt.figure(figsize = (8,8))
plt.pie(sizes, labels=labels, rotatelabels=False, autopct='%1.1f%%',colors=colors,shadow=True, startangle=90)
plt.title('COSTS',color = 'green',fontsize = 15)
plt.show()
```

```python
binary_col = ['four_g','touch_screen','dual_sim']
for i in binary_col:
    ax = sns.countplot(x=i, hue='price_range',data=data,palette='pastel')
    ax.set_xticklabels(["There is not","There is"])
    plt.show()
```
